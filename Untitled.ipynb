```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv(("House_Rent_Dataset.csv"))
display(df)
```

```python
df.info()
```

```python
df = df.drop(["Posted On"],axis = 1)
```

```python
categorical1 = df["Floor"].astype('category')
df["Floor"] = categorical1.cat.codes + 1
```

```python
categorical2 = df["Area Type"].astype('category')
df["Area Type"] = categorical2.cat.codes + 1
```

```python
categorical3 = df["Area Locality"].astype('category')
df["Area Locality"] = categorical3.cat.codes + 1
```

```python
categorical4 = df["City"].astype('category')
df["City"] = categorical4.cat.codes + 1
```

```python
categorical5 = df["Furnishing Status"].astype('category')
df["Furnishing Status"] = categorical5.cat.codes + 1
```

```python
categorical6 = df["Tenant Preferred"].astype('category')
df["Tenant Preferred"] = categorical6.cat.codes + 1
```

```python
categorical7 = df["Point of Contact"].astype('category')
df["Point of Contact"] = categorical7.cat.codes + 1
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(df.corr(), cmap = "Reds", annot = True)
```

```python
df = df.drop(["BHK","Size","Floor","Area Locality","Tenant Preferred","Furnishing Status"],axis=1)
df
```

```python
scalery = StandardScaler()
df["Rent"] = scalery.fit_transform(df["Rent"].to_numpy().reshape(-1, 1))
df
```

```python
scalerx1 = StandardScaler()
df["Area Type"] = scalerx1.fit_transform(df["Area Type"].to_numpy().reshape(-1, 1))
```

```python
scalerx2 = StandardScaler()
df["City"] = scalerx2.fit_transform(df["City"].to_numpy().reshape(-1, 1))
```

```python
scalerx3 = StandardScaler()
df["Bathroom"] = scalerx3.fit_transform(df["Bathroom"].to_numpy().reshape(-1, 1))
```

```python
scalerx4 = StandardScaler()
df["Point of Contact"] = scalerx4.fit_transform(df["Point of Contact"].to_numpy().reshape(-1, 1))
df
```

```python
x = df[["Area Type","City", "Bathroom", "Point of Contact"]]
y = df["Rent"]
mlr = LinearRegression()
mlr.fit(x,y)
```

```python
mlr.coef_
```

```python
mlr.intercept_
```

```python
mlr.predict(x.iloc[4745,:].to_numpy().reshape(1,-1))
```

```python
r2 = r2_score(y, mlr.predict(x))
r2
```
