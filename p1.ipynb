```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(('Netflix Userbase.csv'),sep=',')
display(df)
```

```python
subs = 'Basic', 'Standard', 'Premium'
counts = [999, 768, 733]
plot1 = plt.pie(counts, labels=subs, autopct='%1.1f%%', shadow=True, startangle=90, explode = (0.05,0.0,0.0))
plt.title("Subscription Plans Distribution")
```

```python
df["Age"].value_counts()
```

```python
plot2 = sns.boxplot(df, x = "Age", y = "Gender", linewidth = 1).set(title="Age vs Gender Boxplot")
```

```python
plot3 = sns.countplot(df, x="Device", hue="Gender", palette = "dark:#5A9_r").set(title="Gender vs Device Used")
```

```python
cat1 = {'Basic': 0, 'Standard': 1, 'Premium': 2}
df["Subscription Type"] = df["Subscription Type"].replace(cat1)
df
```

```python
cat2 = {'Male': 0, 'Female': 1}
df["Gender"] = df["Gender"].replace(cat2)
df
```

```python
cat3 = {'Laptop': 0, 'Tablet': 1, 'Smartphone': 2, 'Smart TV': 3}
df["Device"] = df["Device"].replace(cat3)
df
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(df.corr(), cmap = "Reds", annot = True)
```
